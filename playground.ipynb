```python
import matplotlib.pyplot as plt
import numpy as np
import torch

from dataset import PepeDataset
from config import cfg
```

```python
dataset_name = 'pepe'
dataset = PepeDataset(dataset_name, cfg)

sample_num = np.random.randint(len(dataset))
print(f'Sample num is {sample_num}')

sample = dataset[sample_num]
image = sample.numpy().transpose((1, 2, 0))
image = (image + 1) / 2
plt.imshow(image)
plt.show()
```

```python
plt.hist(sample.flatten().numpy(), bins=50)
plt.show()
```
