# Dataset examples

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

from config import Paths, Config
from data.dataset import PepeDataset
from high_resolution_model.highres_dataset import HighResPepeDataset
```

```python
def show_dataset_examples(dataset_name, image_size, grid_size=(3, 5)):
    dataset = PepeDataset(dataset_name, image_size, Paths())
    
    sample_nums = np.random.randint(len(dataset), size=grid_size[0] * grid_size[1])
    print(f'Sample nums are {sample_nums}')
    
    fig, axes = plt.subplots(*grid_size, layout='constrained', figsize=(grid_size[1] * 2, grid_size[0] * 2), squeeze=False)
    for i, sample_num in enumerate(sample_nums):
        sample = dataset[sample_num]
        dataset.show_item(sample, axis=axes[i // grid_size[1], i % grid_size[1]])
    plt.show()
```

```python
dataset_name = 'celeba'
image_size = 128

show_dataset_examples(dataset_name, image_size, grid_size=(5, 1))
```

```python
dataset_name = 'twitch_emotes'
image_size = 128

show_dataset_examples(dataset_name, image_size)
```

```python
dataset_name = 'pepe'
image_size = 128

show_dataset_examples(dataset_name, image_size)
```
